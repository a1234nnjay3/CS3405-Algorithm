```python
class SortMethod(list):
    
    def SelectionSort(self):
        for i in range(0, len(self)-1):
            for j in range(i+1, len(self)):
                if self[i] < self[j]:
                    self[i], self[j] = self[j], self[i]
        return self
    
    def BubbleSort(self):
        if len(self) == 1:
            return self
        
        for i in range(1, len(self)):
            while self[i] > self[i-1] and i>0:
                self[i], self[i-1] = self[i-1], self[i]
                i-=1
        return self
    
    def MergeSort(self):
        if len(self) ==1:
            return self
        
        left = SortMethod(self[:len(self)//2])
        right = SortMethod(self[len(self)//2:])
        
        SortedLeft = left.MergeSort()
        SortedRight = right.MergeSort()
        
        returnList = []
        li = 0
        ri = 0
        while len(SortedLeft) > li and len(SortedRight) > ri:
            if SortedLeft[li] > SortedRight[ri]:
                returnList.append(SortedLeft[li])
                li+=1
            else:
                returnList.append(SortedRight[ri])
                ri+=1
        
        if li == len(SortedLeft):
            returnList.extend(SortedRight[ri:])
        else:
            returnList.extend(SortedLeft[li:])
        return returnList
    
    def HeapSort(self):
        
        def shiftup(self, i):
            if i == 1:
                return
            if self[i] > self[i//2]:
                self[i], self[i//2] = self[i//2], self[i]
                shiftup(self, i//2)
            else:
                return
            
        def shiftdown(self, end, i):
            if end <=2:
                return
            if end < 2*i:
                return
            elif end == 2*i:
                if self[i] < self[2*i]:
                    self[i], self[2*i] = self[2*i], self[i]
                return 
            else:
                if self[i] < max(self[2*i], self[2*i+1]):
                    if self[2*i] > self[2*i+1]:
                        self[i], self[2*i] = self[2*i], self[i]
                        shiftdown(self, end, 2*i)
                    else:
                        self[i], self[2*i+1] = self[2*i+1], self[i]
                        shiftdown(self, end, 2*i+1)
        
        self.insert(0,0)
        for i in range(2, len(self)):
            shiftup(self,i)
        end = len(self)-1
        while self[end] < self[1] and end > 1:
            self[end], self[1] = self[1], self[end]
            end-=1
            shiftdown(self, end, 1)
        self.remove(0)
        return
    
    def QuickSort(self, first, last):
        
        if first >= last:
            return
        pivot = self[first]
        left, right = first+1, last
        while left < right:
            while left < right and self[left] < pivot:
                left+=1
            while right > left and self[right] >= pivot:
                right-=1
            self[left], self[right] = self[right], self[left]
            
        if self[left] < pivot:
            center = left
        else:
            center = left - 1
        
        self[first], self[center] = self[center], self[first]
        self.QuickSort(first, center-1)
        self.QuickSort(center+1, last)
```

```python
import time
import random

random.seed(1)
a100 = SortMethod([ random.randint(0, 9999) for i in range(100)])
start_time = time.clock()
a100.SelectionSort()
print("Run time for Selection Sort is",time.clock() - start_time, "   seconds")

random.seed(1)
a100 = SortMethod([ random.randint(0, 9999) for i in range(100)])
start_time = time.clock()
a100.BubbleSort()
print("Run time for Bubble Sort is   ",time.clock() - start_time, "   seconds")

random.seed(1)
a100 = SortMethod([ random.randint(0, 9999) for i in range(100)])
start_time = time.clock()
a100.MergeSort()
print("Run time for Merge Sort is    ",time.clock() - start_time, "   seconds")

random.seed(1)
a100 = SortMethod([ random.randint(0, 9999) for i in range(100)])
start_time = time.clock()
a100.HeapSort()
print("Run time for Heap Sort is     ",time.clock() - start_time, "   seconds")

random.seed(1)
a100 = SortMethod([ random.randint(0, 9999) for i in range(100)])
start_time = time.clock()
a100.QuickSort(0, len(a100)-1)
print("Run time for Quick Sort is    ",time.clock() - start_time, "   seconds")
```

### Merge sort         :$O(NlogN)$
